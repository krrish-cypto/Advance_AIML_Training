```python
# ---------------------------------------
# 1. Create Sample Dataset (CSV)
# ---------------------------------------
import pandas as pd

data = """customer_id,data_gb,call_mins
1001,5.2,230
1002,12.5,410
1003,7.8,320
1004,15.6,580
1005,3.4,120
1006,9.0,360
1007,2.5,80
1008,11.0,400
1009,18.3,600
1010,4.0,200
"""

# Save dataset to file
open("telecom_usage.csv", "w").write(data)
print("Sample telecom_usage.csv created successfully!")
```

```python
# ---------------------------------------
# 2. Load the Dataset
# ---------------------------------------
import pandas as pd

df = pd.read_csv("telecom_usage.csv")
print("Data Loaded Successfully!")
df.head()
```

```python
# ---------------------------------------
# 3. Import Required Libraries
# ---------------------------------------
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Better plotting style
sns.set(style="whitegrid")
```

```python
# ---------------------------------------
# 4. Apply K-Means Clustering
# ---------------------------------------
kmeans = KMeans(n_clusters=3, random_state=0)

# Train clustering model on 2 columns
kmeans.fit(df[['data_gb', 'call_mins']])

# Add cluster labels to dataframe
df['cluster'] = kmeans.labels_

# Show final clustered data
df
```
